```python
# Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
# Carregamento da base de dados
base = pd.read_csv('orchard.csv')
base
```

```python
#indicando atributo projection = '3d' e passando três atributos (decrease, rowpos e colpos)
figura = plt.figure()
eixo = figura.add_subplot(1, 1, 1, projection = '3d')

grupo1 = eixo.scatter(base['decrease'][:50], base['rowpos'][:50], base['colpos'][:50],
                      c='red', label='Grupo 1', s=50)  # label é essencial para a legenda
grupo2 = eixo.scatter(base['decrease'][50:], base['rowpos'][50:], base['colpos'][50:],
                      c='blue', label='Grupo 2', s=50)

eixo.set_xlabel('decrease')
eixo.set_ylabel('rowpos')
eixo.set_zlabel('colpos')
eixo.legend()
```
