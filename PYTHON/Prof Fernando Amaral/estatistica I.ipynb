Como verificar se é distribuição normal?
Visualização gráfica, teste estatistico, métrica estatistica

Histograma: Plote um histograma dos dados e veja se ele se aproxima de uma curva em formato de sino (a forma típica de uma distribuição normal).

Gráfico Q-Q (Quantil-Quantil): Um gráfico Q-Q compara os quantis dos seus dados com os quantis de uma distribuição normal. Se os pontos estiverem próximos de uma linha reta, isso sugere normalidade.

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
import numpy as np
```

```python
# Dados de exemplo
dados = [10, 12, 14, 15, 15, 16, 17, 18, 20]
# Gráfico Q-Q
stats.probplot(dados, dist="norm", plot=plt)
plt.show()
```

```python
# Histograma
plt.hist(dados, bins=10, edgecolor='k', alpha=0.7)
plt.title("Histograma")
plt.show()
```

Shapiro-Wilk Test: Um dos testes mais comuns. Se o valor-p for maior que 0,05, você não pode rejeitar a hipótese de normalidade.

```python
from scipy.stats import shapiro

# Teste Shapiro-Wilk
stat, p_value = shapiro(dados)
print(f"Estatística do teste: {stat}, Valor-p: {p_value}")

if p_value > 0.05:
    print("Os dados seguem uma distribuição normal (não rejeitamos H0).")
else:
    print("Os dados não seguem uma distribuição normal (rejeitamos H0).")
```

Métricas Estatísticas
Assimetria (Skewness): Em uma distribuição normal, a assimetria deve ser próxima de 0.
Curtose: Deve ser aproximadamente 3 para uma distribuição normal

```python
from scipy.stats import skew, kurtosis

print("Assimetria:", skew(dados))
print("Curtose:", kurtosis(dados, fisher=False))  # Fisher=False para curtose "clássica"
```
