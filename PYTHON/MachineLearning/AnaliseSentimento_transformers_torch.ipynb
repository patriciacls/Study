```python
# modelo VADER
from transformers import pipeline
```

```python
# pipeline com o modelo DistilBERT
analise_sentimento = pipeline("sentiment-analysis", model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
```

```python
frase = input("Digite uma frase para anÃ¡lise de sentimento: ")
```

```python
resultado = analise_sentimento(frase)
```

```python
# Interpretando o resultado
label = resultado[0]['label']  # "POSITIVE" ou "NEGATIVE"
score = resultado[0]['score']  # Probabilidade da previsÃ£o
```

```python
print(f"ðŸ“Š Sentimento detectado: {label} (confianÃ§a: {score:.2f})")
```
